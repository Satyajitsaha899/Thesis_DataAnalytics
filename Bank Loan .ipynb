```python
##import the dataset
df1 = pd.read_csv("S:\Project_Thesis\Jup\export_loan.csv")
df1.head()
```

```python
## import keras
from keras.models import Sequential 
from keras.layers import Dense 
```

```python
import warnings
warnings.filterwarnings("ignore", category=FutureWarning)
```

```python
##random seed for reproduciblity
numpy.random.seed(7)
```

```python
##split the dataset into X and Y
feature_cols = ['loan_id', 'account_id','date','amount','duration','payments']
X = df1[feature_cols] # Features
Y = df1.status # Target variable
```

```python
##create model,add dense layer one by one specifying activation function
model = Sequential()
model.add (Dense(12,input_dim =6, activation='relu')) ## Input layer required input
model.add (Dense(15,activation ='relu'))
model.add (Dense(8,activation ='relu'))
model.add (Dense(10,activation ='relu'))
model.add (Dense(1,activation ='sigmoid'))## sigmoid instead of relu for final prediction
```

```python
##compile the model,adam gredient descent (optimized)
model.compile(loss="binary_crossentropy", optimizer= "adam", metrics = ['accuracy'])
```

```python
## call the function to fit with the dataset (training the network)
model.fit (X,Y, epochs = 1000, batch_size = 10 )
```

```python
## evalute the model
scores = model.evaluate (X, Y)

print ("\n%s: %.2f%%" % (model.metrics_names[1], scores [1] * 100))
```
